```python
#Importando as Bibliotecas
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Lendo o Arquivo com as copetencias
text = open('Copetencias.csv','r',encoding='utf-8').read()
```

```python
#Criando o objeto WordCLoud
wordcloud = WordCloud(max_font_size=100,width = 800, height = 400, background_color='#e6e6ff', collocations=False).generate(text)
#print(wordcloud.words_.items())
#Imprimindo o Grafico Mapa de Palavras
plt.figure(figsize=(19,9))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
